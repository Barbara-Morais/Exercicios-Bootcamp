```python
import pandas as pd
import dtale


import dtale.app as dtale_app

dtale_app.USE_COLAB = True
```

```python
url = 'https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv'
```

```python
import numpy as np

df = pd.read_csv(url, sep=',')
df["Class"] = np.zeros((df.shape[0], 1))
dtale.show(df)
```

```python
df = df.sort_values(by="hp").reset_index().drop('index', axis=1, errors='ignore')
```

```python
df.columns = [str(c) for c in df.columns]  # update columns to strings in case they are numbers

df.at[31, 'Class'] = 1
df.at[30, 'Class'] = 1
df.at[29, 'Class'] = 0
df.at[28, 'Class'] = 1
df.at[27, 'Class'] = 0
df.at[20, 'Class'] = 1
df.at[19, 'Class'] = 1
df.at[18, 'Class'] = 1
df.at[17, 'Class'] = 1
df.at[14, 'Class'] = 1
df.at[5, 'Class'] = 1
df = df.rename(columns={'Class': 'cathegory'})
df = df.sort_values(['hp'], ascending=[True])

df
```

```python
is_sport = df.cathegory == 1
```

```python
print(f"Numero de carros esportivos: {df[is_sport].shape[0]}")
print(f"Numero de carros não esportivos: {df[~is_sport].shape[0]}")
```

```python
dimensions = "mpg	cyl	disp	hp	drat	wt	qsec	carb".split("\t")
X = df[dimensions]
y = df['cathegory']
```

```python
from sklearn.preprocessing import StandardScaler 

X_scaled = pd.DataFrame(StandardScaler().fit_transform(X), columns=X.columns)
```

```python
%matplotlib inline

import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('dark')

fig, axes = plt.subplots(1, 2, figsize=(20, 6))

sns.boxplot(data=X_scaled.loc[is_sport, :], ax=axes[0])
sns.boxplot(data=X_scaled.loc[~is_sport,:], ax=axes[1])

axes[0].grid()
axes[0].set_title('Sport cars')
axes[0].set_xlabel('continuous variables')
axes[0].set_ylabel('scaled values')
axes[1].grid();
axes[1].set_title('Other cars')
axes[1].set_xlabel('continuous variables')
axes[1].set_ylabel('scaled values')
fig.tight_layout()
```

```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.offline as pyo
from plotly.subplots import make_subplots

categories = X_scaled.columns.tolist()

fig_sport = go.Figure(
    data = [
      go.Scatterpolar(r=X_scaled[is_sport].iloc[idx], theta=categories) 
        for idx in range(X_scaled[is_sport].shape[0])
    ],
    layout=go.Layout(polar={'radialaxis': {'visible': True}})
)

fig_nosport = go.Figure(
    data = [
      go.Scatterpolar(r=X_scaled[~is_sport].iloc[idx], theta=categories) 
        for idx in range(X_scaled[~is_sport].shape[0])
    ],
    layout=go.Layout(polar={'radialaxis': {'visible': True}})
)

fig_sport.show()
fig_nosport.show();
```

```python
X[is_sport][0: 4]
```

```python
X_balanced = pd.concat([X, X[is_sport], X[is_sport][0: 4]])
y_balanced = pd.concat([y, y[is_sport], y[is_sport][0: 4]])
```

```python
X_balanced.shape[0]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X_balanced, y_balanced,
                                                    test_size=0.25,
                                                    random_state=171)

scaler = StandardScaler().fit(x_train)

x_train_scaled, x_test_scaled = scaler.transform(x_train), scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

l_regr = LogisticRegression()

l_regr.fit(x_train_scaled, y_train);
```

```python
l_regr.score(x_test_scaled, y_test)
```

```python
predictions = l_regr.predict(x_test_scaled)
predictions
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, predictions)

print(cm)
```
