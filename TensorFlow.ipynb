```python
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras import utils
```

```python
from sklearn import datasets
wine = datasets.load_wine()
print(wine.data.shape)
len(wine.target_names)
```

```python
wine.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.2,random_state=20)
```

```python
y_train = utils.to_categorical(y_train)
y_train
y_test = utils.to_categorical(y_test)
```

```python
model = Sequential()
model.add(Dense(20,input_dim=13,activation='relu'))
#model.add(Dense(7,input_dim=20,activation='relu'))
model.add(Dense(3,input_dim=7,activation='softmax'))
```

```python
model.summary()
```

```python
from tensorflow.keras import optimizers
#hyper-parameters for optimizer
learning_rate = 0.001
opt = optimizers.Adam(lr=learning_rate)

model.compile(loss='categorical_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])
```

```python
y_train.shape
```

```python
from keras.callbacks import EarlyStopping
es = EarlyStopping(monitor='val_loss', mode='min',restore_best_weights=True, verbose=1,patience=10)
```

```python
history = model.fit(X_train,y_train,epochs=300,batch_size=10,validation_split=0.1,callbacks=[es])
```

```python
eval_Train = model.evaluate(X_train,y_train)
eval_Train
```

```python
eval = model.evaluate(X_test,y_test)
```

```python
eval
```

```python
import numpy as np
from sklearn.metrics import multilabel_confusion_matrix
```

```python
import matplotlib.pyplot as plt

training_loss = history.history["loss"]
validation_loss = history.history["val_loss"]
plt.plot(history.epoch, training_loss)
plt.plot(history.epoch, validation_loss)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['training', 'validation'], loc='upper right')
plt.show()
```

# Implementando a rede neural na "m√£o" 

```python
# lendo os pesos da 1a camada #
w1 = model.layers[0].get_weights()[0]
w1.shape
```

```python
# lendo os bias da 1a camada #
b1 = model.layers[0].get_weights()[1]
b1.shape
```

```python
hidden_1 = tf.matmul(X_test,w1) + b1
# multiplica√ß√£o da entrada pela matriz de pesos + vetor de bias 
```

```python
hidden_1.shape
```

```python
out_1 = tf.keras.activations.relu(hidden_1)

# sa√≠da da 1a camada
```

```python
# lendo os pesos da 2a camada (de sa√≠da, neste caso) #
w2 =  model.layers[1].get_weights()[0]
b2 =  model.layers[1].get_weights()[1]
```

```python
hidden_2 = tf.matmul(out_1,w2) + b2
```

```python
out_2 =  tf.keras.activations.softmax(hidden_2)
```
