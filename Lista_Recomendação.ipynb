####1.2 - Quais os filmes com maiores ratings a partir do ano 2000 - salvar em um csv os 1000 mais bem “rankeados”.

```python
movies[movies['year']=='TV Movie 2019']
```

```python
movies['year'].replace(to_replace='TV Movie 2019', value=2019, inplace=True)
```

```python
movies['year'] = movies['year'].astype(int)
```

Para conseguir fazer um critério de ratings numéricos (pelo ano do filme), a coluna que originalmente era "objeto" foi modificada para INT. Contudo, ao tentar realizar a mudança, uma das linhas possuia a informação "TV Movie 2019" - a qual foi modificada individualmente para '2019', de forma a finalmente conseguir realizar a mudança do tipo de toda a coluna.

```python
movies_2000 = movies[movies['year']>=2000]
movies_2000.head()
```

sort_values: com o "ascending = False", a coluna foi ordenada do maior para o menor

```python
movies_rating = movies_2000.sort_values('avg_vote', ascending=False)
movies_rating
```

```python
movies_rating.head(1000).to_csv('maiores_ratings.csv')
```
