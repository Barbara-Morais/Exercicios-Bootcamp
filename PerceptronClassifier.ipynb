```python
class Perceptron_Bootcamp(object):

  def __init__(self,input_size,epochs=50,etha=0.01):
    self.W = np.zeros(1+input_size) 
    #b = 0 
    self.epochs = epochs
    self.etha = etha

  def f_activation(self,x):
   
   return np.where(x>0,1,0)  

  def training(self,X_train,y_train):

    for k in range(self.epochs):
      
      for xi, target in zip(X_train,y_train):
        
        sum = np.dot(xi,self.W[1:]) + self.W[0]
        output = self.f_activation(sum)
        err = target - output
        update = self.etha*(err)
        self.W[1:] += update*xi
        self.W[0] += update*1
    return self.W    
  
  def predict(self,X_test):
    sum = np.dot(X_test,self.W[1:]) + self.W[0]
    output = self.f_activation(sum)
    return output
```

```python
tico_Perceptron = Perceptron_Bootcamp(X_train.shape[1],etha=1.0)
```

```python
w = tico_Perceptron.training(X_train,y_train)
```

```python
y_pred = tico_Perceptron.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
